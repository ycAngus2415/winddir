```python
for filter_ in tqdm(os.listdir('/Volumes/Yangchao/speckle/')):
    sigma_vh =[]
    sigma_vv =[]
    incidence = []
    scending = []
    wspeed = []
    wdir = []
    height = []
    sar_data={}
    for file in tqdm(os.listdir('/Volumes/Yangchao/speckle/'+filter_)):
        dataset = nc.Dataset('/Volumes/Yangchao/speckle/'+filter_+'/'+file)
        if dataset.variables['Sigma0_VH'][:].shape!=(400, 400):
            dataset.close()
            del dataset
            gc.collect()
            continue
        
        sigma_vh.append(dataset.variables['Sigma0_VH'][:])
        sigma_vv.append(dataset.variables['Sigma0_VV'][:])
        incidence.append(dataset.variables['incident_angle'][:])
        file_list = file.split('_')
        scending.append(file_list[0])
        wdir.append(int(file_list[3]))
        wspeed.append(float(file_list[4]))
        height.append(float(file_list[5]))
        dataset.close()
        del dataset
        gc.collect()
        
#     with open('/Volumes/Yangchao/sigma_vv'+'_'+filter_, 'wb') as v:
#         pickle.dump(sigma_vv, v)
#     with open('/Volumes/Yangchao/sigma_vh'+'_'+filter_, 'wb') as h:
#         pickle.dump(sigma_vh, h)
#     with open('/Volumes/Yangchao/incidence'+'_'+filter_, 'wb') as i:
#         pickle.dump(incidence, i)


    datas = pd.DataFrame({'wspeed':wspeed, 'wdir':wdir, 'height':height, 'scending':scending})
    datas.to_csv('/Volumes/Yangchao/'+filter_, index=False)
    sar_data['sigma_vv']=sigma_vv
    sar_data['sigma_vh']=sigma_vh
    sar_data['incidence']=incidence
    with open('/Volumes/Yangchao/sar_'+filter_, 'wb') as ss:
        pickle.dump(sar_data, ss)
    gc.collect()

```
