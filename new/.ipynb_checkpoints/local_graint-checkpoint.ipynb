```python
def smooth(img):
    B_x = np.array([1, 1])/2
    B_x2 = np.array([1, 2, 1])/4
    B_y2 = np.array([[1],[2],[1]])/4
    B_x_y2 = np.array([[0,0,1],[0,2,0],[1,0,0]])/4
    B_x__y2 = np.array([[1,0,0],[0,2,0],[0,0,1]])/4
    B2 = B_y2*B_x2
    x = np.array([1, 4, 6, 4, 1])/16
    y = np.reshape(x, (5, 1))
    B4 = y*x
    conv1 = signal.convolve2d(img, B2, mode='valid')
    conv2 = signal.convolve2d(conv1, B4, mode='valid')
    return conv2
```

```python
def sobel(img):
    r = np.array([[3,0,-3],[10,0,-10],[3,0,-3]])/32
    sob1 = signal.convolve2d(img, r,mode='valid')
    sob2 = signal.convolve2d(img, r.T,mode='valid')
    return sob1, sob2
```

```python
def wind_direction(vv):
    bins1=np.arange(19)*10
    im_smooth = smooth(vv)
    im_ssobelx, im_ssobely = sobel(im_smooth)
    anotention = im_ssobely/im_ssobelx
    azimuth = np.arctan(anotention)/np.pi*180
    azimuth1 = azimuth%180
    his,bins2 = np.histogram(azimuth1, bins=bins1,range=(0,180))
    direction = np.argmax(his)*10+5
    del bins2
    del bins1
    return direction
```

```python
def ambugity(dire_, dire_ndbc):
    if dire_-dire_ndbc>0:
        if (dire_ndbc-dire_)%180<dire_-dire_ndbc:
            return (dire_+180)%360, (dire_ndbc-dire_)%180
        else:
            return dire_, dire_-dire_ndbc
    else:
        if (dire_-dire_ndbc)%180<dire_ndbc-dire_:
            return (dire_+180)%360, (dire_-dire_ndbc)%180
        else:
            return dire_, dire_ndbc-dire_
```

```python
from tqdm import tqdm
```

```python
t = []
error = []

for s_vv,dir1 in tqdm(zip(sigma_vv,data['wdir'])):
    direction1 = wind_direction(s_vv)
    d, e= ambugity(direction1, dir1)
    t.append(d)
    error.append(e)
```
